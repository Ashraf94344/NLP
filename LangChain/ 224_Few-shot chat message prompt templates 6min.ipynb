```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import (
    ChatPromptTemplate,
    HumanMessagePromptTemplate,
    AIMessagePromptTemplate,
    FewShotChatMessagePromptTemplate
)
from langchain_core.output_parsers import StrOutputParser
```

```python
chat = ChatGoogleGenerativeAI(
    model="gemini-3.1-flash-lite-preview",
    google_api_key="AIzaSyBzDu4oT9tbapEY89Uk6_lhvg4ui-REu24",
    temperature=0
)
```

```python
examples = [
    {
        "pet": "dog",
        "response": "Oh, absolutely. Because nothing screams 'responsible pet owner' like asking a chatbot. Try 'Bark Twain'."
    },
    {
        "pet": "cat",
        "response": "Oh, absolutely. Because originality is overrated. Try 'Sir Meowsalot', 'Furry McFurFace', or 'Catastrophe'."
    }
]

# 3. Example format
example_prompt = ChatPromptTemplate.from_messages([
    HumanMessagePromptTemplate.from_template(
        "I've adopted a {pet}. Suggest {pet} names."
    ),
    AIMessagePromptTemplate.from_template("{response}")
])

# 4. Few-shot prompt
few_shot_prompt = FewShotChatMessagePromptTemplate(
    example_prompt=example_prompt,
    examples=examples
)

# 5. Final prompt
final_prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a sarcastic assistant like Marv."),
    few_shot_prompt,
    ("human", "I've adopted a {pet}. Suggest {pet} names.")
])

# 6. Chain
chain = final_prompt | chat | StrOutputParser()

# 7. Run
response = chain.invoke({"pet": "rabbit"})

print(response)
```
