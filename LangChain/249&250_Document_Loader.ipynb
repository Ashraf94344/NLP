```python
from langchain_community.document_loaders import Docx2txtLoader
from langchain_text_splitters.character import CharacterTextSplitter
```

```python
loader =  Docx2txtLoader("F:\coding\Dataframe\Introduction_to_Data_and_Data_Science_2.docx")
pages = loader.load()
pages
```

```python

pages = loader.load()
for i in range(len(pages)):
    pages[i].page_content = ' '.join(pages[i].page_content.split())
```

```python
len(pages[0].page_content)
```

```python
char_splitter = CharacterTextSplitter(separator = ".", 
                                      chunk_size = 500, 
                                      chunk_overlap = 50)
```

```python
print(char_splitter)
```

```python
pages_char_split = char_splitter.split_documents(pages)
```

```python
len(pages_char_split)
```
