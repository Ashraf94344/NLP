```python
response = model.generate_content(
    "Could you explain briefly what a black hole is?",
    generation_config={
        "max_output_tokens": 250,
        "temperature": 0
    },
    stream=True   ###output will come part by part
)

for i in response:
    print(i.text, end="")
```
